```python
from google_play_scraper import app

import pandas as pd

import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.openai.chatgpt',
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT you can use Sort.NEWEST to get newst reviews
    count=800, # defaults to 1000
    filter_score_with=None # defaults to None(means all score) Use 1 or 2 or 3 or 4 or 5 to select certain score
)
```

```python
df_gpt = pd.DataFrame(np.array(result),columns=['review'])

df_gpt = df_gpt.join(pd.DataFrame(df_gpt.pop('review').tolist()))

df_gpt.head()
```

```python
df_gpt.to_csv("scrapped_gpt.csv", index = False)  #save to csv
```

# 3. Labelling

```python
data = pd.read_csv('/content/scrapped2.csv')
data.head()
```

```python
def pelabelan(score):
  if score < 3:
    return 'Negatif'
  elif score == 3 :
    return 'Netral'
  elif score == 5 or score == 4 :
    return 'Positif'
data['label'] = data ['score'].apply(pelabelan)
data.head()
```

```python
data.to_csv("scrapped_label.csv", index = False)  #save to csv
```
